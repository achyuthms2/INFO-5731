# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
# removing special characters

import re
reviews['AfterProcessing'] = reviews['review'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
reviews

```

```python
# removing punctuations

reviews['AfterProcessing'] = reviews['AfterProcessing'].str.replace('[^\w\s]','')
reviews
```

```python
# removing numbers

reviews['AfterProcessing'] = reviews['AfterProcessing'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
reviews
```

```python
# removing stopwords

import nltk
nltk.download()
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
reviews['AfterProcessing'] = reviews['AfterProcessing'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
reviews
```

```python
# lowercase

reviews['AfterProcessing'] = reviews['AfterProcessing'].apply(lambda x: " ".join(x.lower() for x in x.split()))
reviews
```

```python
# lemmatization

from textblob import Word
import nltk
nltk.download('wordnet')

reviews['AfterProcessing'] = reviews['AfterProcessing'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
reviews
```

```python
# stemming

from nltk.stem import PorterStemmer
st = PorterStemmer()
reviews['AfterProcessing'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

(30 points). Write a python program to conduct **syntax and structure analysis** of the clean text you just saved above. The syntax and structure analysis includes: 

(1) Parts of Speech (POS) Tagging: Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) Constituency Parsing and Dependency Parsing: print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) Named Entity Recognition: Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
# Write your code here
from nltk.tokenize import word_tokenize
pos = []
for sentence in reviews['AfterProcessing']:
  text = word_tokenize(sentence)
  pos.append(nltk.pos_tag(text))
pos
```

```python
Noun = []
Verb = []
Adjective = []
Adverb = []
for value in pos:
  NounCount = 0
  VerbCount = 0
  AdjectiveCount = 0
  AdverbCount = 0
  for word,tag in value:
    if tag == 'NN':
      NounCount = NounCount + 1
    elif tag == 'VR':
      VerbCount = VerbCount + 1
    elif tag == 'JJ':
      AdjectiveCount = AdjectiveCount + 1
    elif tag == 'RB':
      AdverbCount = AdverbCount + 1
  Noun.append(NounCount)
  Verb.append(VerbCount)
  Adjective.append(AdjectiveCount)
  Adverb.append(AdverbCount)
reviews['Number_of_Nouns'] = Noun
reviews['Number_of_Verbs'] = Verb
reviews['Number_of_Adjectives'] = Adjective
reviews['Number_of_Adverbs'] = Adverb
reviews
```
